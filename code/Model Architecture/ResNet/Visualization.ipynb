```python
# imports
import keras
import os
import tensorflow as tf
import visualkeras
from PIL import ImageFont
from keras.utils import plot_model
import pydot, graphviz
```

```python
def draw_model(model_name, save = True):

    # Setup Paths
    model_path = os.path.join("models_ResNet", model_name)
    output_path = os.path.join("visualizations", f"{model_name.split(".")[0]}.png")
    loaded_model = keras.saving.load_model(model_path)
    
    # Set Font
    font = ImageFont.truetype("Arial.ttf", 50)

    # Params
    scale_xy = 5
    max_z = 25
    spacing = 20
    legend = True
    
    # Generate Image
    if save:
        visualkeras.layered_view(loaded_model, scale_xy = scale_xy, max_z = max_z, spacing = spacing, legend=legend, font=font, to_file=output_path).show()
    else:
        visualkeras.layered_view(loaded_model, scale_xy = scale_xy, max_z = max_z, spacing = spacing, legend=legend, font=font).show()

def graph_model(model_name, save = False):

    # Setup Paths
    model_path = os.path.join("models_ResNet", model_name)
    output_path = os.path.join("visualizations", f"{model_name.split(".")[0]}_graph.png")
    loaded_model = keras.saving.load_model(model_path)

    
    # Generate Image
    if save:
        plot_model(loaded_model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
    else:
        plot_model(loaded_model, show_shapes=True, show_layer_names=True)
```

```python
draw_model("model_ResNet18_3_dropout_callbacks_sigmoid_biological_channels_neurons_trained.keras")
```
